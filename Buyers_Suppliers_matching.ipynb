```
import pandas as pd;

buyerSupplierData = pd.read_csv('matchings.csv')
buyerSupplierData.head

#Get buyers and suppliers list
buyerList = list(set(buyerSupplierData["BuyerType"].tolist()))
supplierList = list(set(buyerSupplierData["SupplierType"].tolist()))

#Get  count of buyers
buyerCount = len(set(buyerSupplierData["SupplierType"].tolist()))

#Create an empty data frame to store supplier affinity scores for suppliers
supplierAffinity = pd.DataFrame(columns=('supplier1', 'supplier2', 'score'))
rowCount = 0

for ind1 in range(len(supplierList)):
  #Get list of buyers who bought from this supplier1
  supplier1Buyers = buyerSupplierData[buyerSupplierData.SupplierType == supplierList[ind1]]["BuyerType"].tolist()
  
  #print("Supplier 1 ", supplier1Buyers)
  
  #Get supplier 2 - suppliers
  for ind2 in range(ind1, len(supplierList)):
    if(ind1 == ind2):
      continue
      
    #Get list of buyers who bought from supplier 2
    supplier2Buyers = buyerSupplierData[buyerSupplierData.SupplierType == supplierList[ind2]]["BuyerType"].tolist()
  
    #print("Supplier  ", supplier2Buyers)
    
    #Find score
    commonBuyers = len(set(supplier1Buyers).intersection(set(supplier2Buyers)))
    score = commonBuyers / buyerCount
    
    #Add a score for supplier1, supplier2
    supplierAffinity.loc[rowCount] = [supplierList[ind1], supplierList[ind2], score]
    rowCount += 1
    
    #Add a score for supplier2, supplier1
    #supplierAffinity.loc[rowCount] = [supplierList[ind2], supplierList[ind1], score]
    rowCount += 1
    
    #print result
    supplierAffinity.head()
    print(supplierAffinity)
    
    

```

Here, we try to find the affinity between a type of buyer and some types of suppliers. The types represents just the data of the buyers or suppliers (location, product types, purchase history, etc.) and the requirements they defined (certifications needed, transportation means, delivery time, products price range, etc.).  

It is preferable to use this approach to find types of users and then find the affinity between these types instead of trying to match buyers and suppliers on each attribute (by location, price, etc.). The approach used is somehow similar to a PCA reduction, used in machine learning.

```
searchSupplier = 5001

recList = supplierAffinity[supplierAffinity.supplier1==searchSupplier][['supplier2', 'score']].sort_values('score', ascending=[0])
print(recList)
```
